```python
y = df.label.values
x_data = df.drop(['label'],axis =1)
```

```python
x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data)).values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 122580)
```

```python
print("x_train.shape : ", x_train.shape)
print("x_test.shape : ", x_test.shape)
print("y_train.shape : ", y_train.shape)
print("y_test.shape : ", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score, cohen_kappa_score

logreg = LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
f=logreg.predict(x_test)
accscore = accuracy_score(y_test, f)
recscore = recall_score(y_test, f)
f1score = f1_score (y_test, f)
kappascore =cohen_kappa_score(y_test,f)
precision = precision_score(y_test, f)


print('Accuracy Score: ',accscore,'\n')
print('Recall Score: ',recscore,'\n')
print('F1 Score: ',f1score,'\n')
print('Kappa Score: ',kappascore,'\n')
print('Precision Score: ',precision,'\n')
```

```python
from sklearn.metrics import roc_curve

lr_probs = logreg.predict_proba(x_test)
lr_probs = lr_probs[:, 1]

fpr, tpr, thresholds = roc_curve(y_test, lr_probs)
print('Fpr: ',fpr,'\n')
print('Tpr: ',tpr,'\n')
print('Thresholds: ',thresholds,'\n')
```

```python
plt.figure(figsize = (12,6))
plt.plot(fpr,tpr,marker = '.',color = 'black')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.grid('on')
```

```python
from sklearn.metrics import precision_recall_curve

precision, recall, thresholds = precision_recall_curve(y_test, lr_probs)

f1_var =[]
for i in range(len(recall)):
    f1 = 2*(precision[i]*recall[i])/(precision[i]+recall[i])
    f1_var.append(f1)
    
```

```python
mx = np.max(f1_var[: -1])
mxmx = [mx for x in range(len(thresholds)) ]
idx = f1_var.index(mx)
print(thresholds[idx])
print(f1_var[idx])
trtr = [thresholds[idx] for x in range(len(thresholds)) ]



plt.figure(figsize=(12,6))

plt.plot(thresholds[:idx],mxmx[:idx],"r--")
plt.plot(trtr,f1_var[: -1],"r--")


plt.plot(thresholds,f1_var[: -1], color='darkblue',linewidth=2)
plt.ylabel("F1-Score")
plt.xlabel("Threshold")
plt.grid('on')
```
